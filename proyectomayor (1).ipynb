```python
 data_csv.loc[:,["Country","Year","Average_Temperature_C"]]
```

### Temperatura más alta del registro.

```python
valor_maximo = data_csv["Average_Temperature_C"].max()
print("El valor máximo de temperatura es:", valor_maximo)
```

```python
data_csv["Average_Temperature_C"] = np.max(data_csv["Average_Temperature_C"])
data_csv["Average_Temperature_C"]
```

### temperatura más baja del registro.

```python
valor_min = data_csv["Average_Temperature_C"].min()
print("El valor minimo de temperatura es:", valor_minimo)
```

```python
data_csv["Average_Temperature_C"] = np.min(data_csv["Average_Temperature_C"])
data_csv["Average_Temperature_C"]
```

### Mirar si hay algo fuera de lo cumún en la Average_Temperature_C, ya que el resultado maximo y el minimo es el mismo (posible error).

```python
data_csv["Average_Temperature_C"] = plt.boxplot(data_csv["Average_Temperature_C"].dropna())
plt.title("Boxplot (para detectar outliers)")
plt.show()
```

### Mirar si hay algo fuera de lo cumún en la columna Soil_Health_Index, (posible error).

### gráfico de dispersión

```python
x =data_csv["Year"]
y =data_csv["Economic_Impact_Million_USD"] 
# Crear gráfico de dispersión cuantitativa discreta y continua
plt.scatter(x, y, color='red') 
# Agregar título y etiquetas
plt.title('Gráfico de Dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y') 
# Mostrar el gráfico
plt.show() 
```

### gráfico de barras

```python
categorias = data_csv["Adaptation_Strategies"]
valores = data_csv["Irrigation_Access_%"] 
# Crear gráfico de barras cualitativas o cuantitativas discretas 
plt.bar(categorias, valores, color='skyblue') 
# Agregar título y etiquetas
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores') 
# Mostrar el gráfico
plt.show()
```

### histograma

```python
datos = ["Total_Precipitation_mm"]
# Crear histograma cuantitativa continua
plt.hist(datos, bins=5, color='purple', edgecolor='black')
# Agregar título y etiquetas
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia') 
# Mostrar el gráfico
plt.show() 
```
