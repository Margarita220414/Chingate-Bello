paso_5:Indexacion y segmentacion (slicing)

1_Acceder a elementos:

```python
arreglo_2d=np.array([[1, 2, 3], [4, 5, 6]])
print(arreglo_2d[:, 1]) # elemento en fila 0, columna 1:2
```

2_segmentación:

```python
print(arreglo_2d[:, 1]) # columna 1: [2 5]
print(arreglo_2d[1, :]) # fila 1:[4 5 6]
```

3_Condiciones (indexación booleana):

```python
print(arreglo_2d[arreglo_2d > 3]) # elementos mayores a 3:[4 5 6]
```

paso_6: Cambiar la forma de los arreglos

1_Redimensionar:

```python
arreglo= np.arange(6) # [0 1 2 3 4 5]
nuevo= arreglo.reshape(2, 3)
print(nuevo)
```

2_Aplanar:

```python
plano=nuevo.flatten()
print(plano) # [0 1 2 3 4 5]
```

paso_7: Operaciones avanzadas

1_Producto punto:

```python
a= np.array([[1, 2], [3, 4]])
b= np.array([[5, 6], [7, 8]])
print(np.dot(a, b))
```

salida:
[[19 22]
[43 50]]

2_Transpuesta:

```python
print(a.T)
```

salida:
[[1 3]
[2 4]]

3_Estadísticas:

```python
print("suma:", np.sum(a)) # 10
print("promedio:", np.mean(a)) # 2.5
print("maximo:", np.max(a)) # 4
```
