continuamos con el tema de estadistica

```python
import random
from typing import Counter


datos = [random.randint(1,5) for _ in range(3)] # lista random

frecuencia_absoluta= Counter (datos) # frecuencia absoluta

valores_ordenados = sorted(frecuencia_absoluta.keys()) # ordena los valores para calcular la frecuencia acomulada

total_datos = len(datos) # calcular la frecuencia acomulada
tabla_frecuencia = [
    (valor,
     frecuencia_absoluta[valor],
     frecuencia_absoluta[valor] / total_datos) # frecuencia relativa
     for valor in valores_ordenados
]

frecuencia_acomulada = []
acomulado = 0
for valor,frec_abso, frec_rela in tabla_frecuencia :
    acomulado += frec_abso
    frecuencia_acomulada.append ((valor, frec_abso, frec_rela, acomulado))

    for valor, frec_abso, frec_rela, frec_acom in frecuencia_acomulada:
        print(f"{'Valor':<10}{'Frec Abso.':<15}{'Frec Rela.':<18}{'Frec Acom.'}")

    for valor, frec_abso, frec_rela, frec_acom in frecuencia_acomulada:
           print(f"{valor:<10}{frec_abso:<15}{frec_rela:<18.2f}{frec_acom}")
```
